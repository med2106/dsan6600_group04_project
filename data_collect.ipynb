# Data Collection and Cleaning Flow

```python
import pandas as pd
import pinscrape
import os
```

## General Flow of Project

1. Collect images from pinterest with pinscrape package

    Pinscrape documentation:
    https://github.com/iamatulsingh/pinterest-image-scrap/tree/master

2. Pass images through YOLO to makes sure it is just one person per image

3. Pass through a facial recognition CNN to make sure it is the back of a person's head

4. Use finished data set to train model

5. Build CNN architecture with PyTorch

6. Pickle model

7. Make light-weight frontend to take image and make curl type prediction

8. Deploy to HuggingFace



```python
from pinscrape import scraper, Pinterest


keyword = "straight+hair"
output_folder = f"data/raw/{keyword}"
proxies = {}
number_of_workers = 10
images_to_download = 10

if not os.path.exists(output_folder):
    os.mkdir(output_folder)

def using_pinterest_apis():
    p = Pinterest(proxies=proxies, sleep_time=2) # you can also pass `user_agent` here.
    images_url = p.search(keyword, images_to_download)
    p.download(url_list=images_url, number_of_workers=number_of_workers, output_folder=output_folder)
    print(images_url)

using_pinterest_apis()
```
